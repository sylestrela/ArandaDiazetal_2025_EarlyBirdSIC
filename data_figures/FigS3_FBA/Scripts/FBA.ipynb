```python
import cobra
from CAFBAFY import * 
import pandas as pd
```

# Load the Data

```python
%%capture
### For universal model ###
model = CAFBA_Model(cobra.io.read_sbml_model('../Data/universe_bacteria.xml'))
### For E.coli model ###
model2 = CAFBA_Model(cobra.io.read_sbml_model('../Data/iJO1366.xml'))


model.set_minimal_media()
model2.set_minimal_media()
model.reactions.EX_glc__D_e.lower_bound=-0.0
model2.reactions.EX_glc__D_e.lower_bound=-0.0
model.reactions.EX_o2_e.lower_bound=-0.0
model2.reactions.EX_o2_e.lower_bound=-0.0
```

```python
#E.coli model lacks oxaloacetate transporter so we have to add in manually (it exists just the metabolic network
# reconstruction
model2.add_reactions([model.reactions.EX_oaa_e.copy(),
                      model.reactions.OAAt2_2pp.copy(),
                      model.reactions.OAAtex.copy()])
```

```python
Csources = ['lac__L_e','cit_e','succ_e','fum_e','oaa_e','pyr_e','ac_e','glyc_e','galt_e',
            'fuc__L_e','rmn_e','mnl_e','sbt__D_e','rib__D_e','arab__L_e',
            'man_e','fru_e','sucr_e','gal_e','malt_e','glc__D_e','lcts_e','melib_e',]
Cmoles = [3,6,4,4,4,3,2,3,6,6,6,6,6,5,5,6,6,12,6,12,6,12,12]
```

# Anaerobic CSS measure growth rates

```python
# List of carbon sources

#Remove ATP Constraints so that result is concentration indepndent.
model.reactions.ATPM.lower_bound=0.0
model2.reactions.ATPM.lower_bound=0.0
# For each carobn sources and each mdoel we are going to calculate optimal growth with a vmax =1 cmo,s. 
fba_g =[]
names=[]
for i in range(0,len(Csources)):
    names.append(model.metabolites.get_by_id(Csources[i]).name)
    #for e.coli
    with model2 as m:
        m.reactions.get_by_id('EX_' + Csources[i]).lower_bound=-1.0/Cmoles[i]
        fba_g.append(m.optimize().objective_value)
        m.reactions.get_by_id('EX_' + Csources[i]).lower_bound=0
        
```

```python
results = pd.DataFrame({'Carbon_Source': names,
             'Pred_Growth_Rate':fba_g,
            'Cmoles': Cmoles})
results.to_csv('../Data/FBA_Growth_Rates.csv')
```
